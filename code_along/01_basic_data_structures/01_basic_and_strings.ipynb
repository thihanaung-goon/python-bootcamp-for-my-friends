## Print Formatting

.format() method just to know

```python
print("This is a string {}".format('ADDED WORD'))
```

```python
print("London {} is {} {}".format("bridge", "fallen", "down"))
```

```python
print("London {0} is {0} {2}".format("bridge", "fallen", "down"))
```

```python
print("London {a} is {b} {c}".format(a="bridge", b="fallen", c="down"))
```

**f-strings**

```python
name = "London"
happen = "fallen down"
print(f"{name} bridge is {happen}")
```

```python
name = "John"
age = 30
print(f"{name} is {age} year old.")
```

```python
result = 10/3
print(result)
```

```python
print(f"{result:.3f}")
```
