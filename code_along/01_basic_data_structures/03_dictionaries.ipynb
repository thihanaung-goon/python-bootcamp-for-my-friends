## Nested dictionaries

```python
students = {
    'Mg Mg': {
        'age': 20,
        'major': 'Physics',
        'scores': [67, 76, 80]
    },
    'Aung Aung': {
        'age': 22,
        'major': 'Chemistry', 
        'scores': [55, 77, 89]
    }
}

students
```

```python
students['Mg Mg']
```

```python
students['Mg Mg']['age']
```

```python
students['Mg Mg']['scores'][-1]
```
